```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# encoding is given as unicode_escape to avoid any errors when we load the data
df = pd.read_csv('Sales Data.csv', encoding='unicode_escape')
```

```python
# to check if the data is loaded properly
# also tells the number of rows and colums in the file
df.shape
```

```python
# to check all the column names and get first 10 rows
df.head(10)
```

```python
# now to start with data cleaning we will get relevant information of the dataset
df.info()
```

```python
# According to the dataset info we will drop null/unwanted columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
# now we will search for null values in the dataset and identify them to remove them
pd.isnull(df).sum()
```

```python
# we have 12 null values in 'Amount' column which we will now remove
# checking number of rows before removing them
df.shape
```

```python
# 11251 before removing, now we will drop the null values
df.dropna(inplace=True)
```

```python
# checking number of rows after dropping null values
df.shape
```

```python
# it is 11239 which means we dropped 12 rows with null value in 'Amount' column
# checking for null values to confirm
pd.isnull(df).sum()
```

```python
# all colums have 0 null values
# now we want to change the data type of the 'Amount' column from float to integer

df['Amount'] = df['Amount'].astype('int')
```

```python
# to check if the datatype changed to integer or not
df['Amount'].dtypes
```

```python
# now we will get the list of the colums in the dataset
df.columns
```

```python
# we will change the column name 'Cust_name' to 'Customer_name'
df.rename(columns = {'Cust_name' : 'Customer_name'})
```

```python
# here we will get description of selected columns in the data frames like mean, count, maximun value and etc.
df[['Age', 'Orders', 'Amount']].describe()
```

```python
# NOW DOING EXPLORATORY DATA ANALYSIS

# BASED ON GENDER HOW MANY CUSTOMERS ARE FEMALE AND HOW MANY ARE MALE WITH VISUAL DATA REPRESENTATION

ax = sns.countplot(x='Gender', hue='Gender', legend=False, palette='bwr', saturation=0.8, data=df)

# now we will label the data values of the graph

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Now we will see the total amount spent by each gender with visual representation using groupby
# we took sum of amount spent by both males and females from the main data frame and saved it in sales_gender as new data frame

sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# here we can see the sales_gender dataframe
sales_gender
```

```python
# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.barplot(x='Gender', y='Amount', data=sales_gender, hue='Gender', saturation=0.8, palette='icefire')
```

```python
'''From both the graphs above we can conclude most of the buyers are female and 
the total amount spent by women is also more which means most products are bought by women'''
```

```python
# Now we will build a count plot for each age group and how many people in each gender are in those age groups

ay = sns.countplot(x='Age Group', hue='Gender', legend=False, palette='bwr', saturation=0.8, data=df)

for bars in ay.containers:
    ay.bar_label(bars)
```

```python


# Now we will see the total amount spent by each age group with visual representation using groupby
# we took sum of amount spent by each age group from the main data frame and saved it in sales_age as new data frame

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', saturation=0.8, palette='icefire')
```

```python
'''From the graphs above we can conclude that females of age group 26-35 are the people who bought the most of the products'''
```

```python
# Now we will see the top 10 states with the most amount of orders
# similarly like above graph we will use group by to create new data frame based on state and number of orders

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# now we have to set size of the plot to avoid overlapping of state names
sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="darkgrid")
sns.barplot(x='State', y='Orders', data=sales_state, hue='State', saturation=0.8, palette='icefire')
```

```python
# Now we will see the top 10 states with the most amount or sales
# similarly like above graph we will use group by to create new data frame based on state and amount

sales_state2 = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# now we have to set size of the plot to avoid overlapping of state names
sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="darkgrid")
sns.barplot(x='State', y='Amount', data=sales_state2, hue='State', saturation=0.8, palette='icefire')
```

```python
'''Here we can conclude with the help of graphs that the most number of orders and most amount of sales is generated from the state Uttar Pradesh'''
```

```python
# Now we will see how many people are married and how many are not married

az = sns.countplot(x='Marital_Status', hue='Marital_Status', legend=False, palette='bwr', saturation=0.8, data=df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in az.containers:
    az.bar_label(bars)
```

```python
# Now we will see the total amount spent based on gender and Marital_Status with visual representation using groupby
# we took sum of amount spent by each gender of each Marital_Status from the main data frame and saved it in sales_m as new data frame

sales_m = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.barplot(x='Marital_Status', y='Amount', data=sales_m, hue='Gender', saturation=0.8, palette='icefire')
```

```python
'''From above graphs we can conclude that married women have purchased the most amount of products'''
```

```python
# Now we will build a count pot for each type of occupation


a1 = sns.countplot(x='Occupation', hue='Occupation', legend=False, palette='bwr', saturation=0.8, data=df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in a1.containers:
    a1.bar_label(bars)
```

```python
# Now we will see the total amount spent based on occupation with visual representation using groupby
# we took sum of amount spent by each occupation from the main data frame and saved it in sales_o as new data frame

sales_o = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_o, hue='Occupation', saturation=0.8, palette='icefire')
```

```python
'''From above graphs we can conclude that most of the buyers are working in IT industry, Healthcare, and Healthcare'''
```

```python
# Now in product category we will build countplot based on sales of each type of product

sns.set(rc={'figure.figsize':(20,5)})
a2 = sns.countplot(x='Product_Category', hue='Product_Category', legend=False, palette='bwr', saturation=0.8, data=df)
for bars in a2.containers:
    a2.bar_label(bars)
```

```python
# Now we will see the total amount based on each type of product with visual representation using groupby
# we took sum of total amout of sales of each type of product from the main data frame and saved it in sales_p as new data frame

sales_p = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data=sales_p, hue='Product_Category', saturation=0.8, palette='icefire')
```

```python
'''From above graphs we can conclude that the most sold product categories are Food, Clothing, and Electronics'''
```

```python
# Now we will plot the graph for top 10 most sold products

sales_top = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Now we will plot a graph chart for this dataframe

sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID', y='Orders', data=sales_top, hue='Product_ID', saturation=0.8, palette='icefire')
```

```python
# CONCLUSION :
'''We can conclude that the most sold product categories are Food, Clothing, and Electronics. 
Also, most of the buyers are working in IT industry, Healthcare, and Healthcare. 
We can also conclude that married women have purchased the most amount of products and 
females of age group 26-35 are the people who bought the most of the products.
The most amount of sales is generated from the state Uttar Pradesh. 
'''
```
